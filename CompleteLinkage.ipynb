```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel("/content/skor.xlsx")
print (data)
```

```python
x = np.asarray(data)
print (x)
```

```python
Z = linkage(x, 'complete')
print(Z)
```

```python
plt.figure(figsize=(25, 10))
plt.title('Dandogram Hierarchical Klastering')
plt.xlabel('data')
plt.ylabel('jarak')
dendrogram(Z)
plt.show()
```
